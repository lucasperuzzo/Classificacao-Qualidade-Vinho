### Informações sobre o projeto

Dataset (conjunto de dados a ser utilizado) está disponível em:
    https://archive.ics.uci.edu/ml/datasets/Wine+Quality

Data Set Information:

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine. For more details, consult: [Web Link] or the reference [Cortez et al., 2009]. Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).

These datasets can be viewed as classification or regression tasks. The classes are ordered and not balanced (e.g. there are many more normal wines than excellent or poor ones). Outlier detection algorithms could be used to detect the few excellent or poor wines. Also, we are not sure if all input variables are relevant. So it could be interesting to test feature selection methods.


Attribute Information:

Input variables (based on physicochemical tests): <br>
- 1 - fixed acidity
- 2 - volatile acidity
- 3 - citric acid
- 4 - residual sugar
- 5 - chlorides
- 6 - free sulfur dioxide
- 7 - total sulfur dioxide
- 8 - density
- 9 - pH
- 10 - sulphates
- 11 - alcohol

Output variable (based on sensory data): <br>
- 12 - quality (score between 0 and 10)

# ETAPA 1

## MEDIDAS DE POSIÇÃO

```python
#Média
media_vinhos = notas_vinhos.groupby(by='Tipo').mean() 
media_vinhos.loc['Total'] = notas_vinhos.mean()
media_vinhos
```

```python
#Mediana
mediana_vinhos = notas_vinhos.groupby(by='Tipo').median()
mediana_vinhos.loc['Total'] = notas_vinhos.median()
mediana_vinhos
```

```python
#Moda
moda_vinhos = notas_vinhos.groupby(by='Tipo').agg(lambda x:x.value_counts().index[0])
moda_vinhos.loc['Total'] = notas_vinhos.mode().loc[0]
moda_vinhos
```

```python
#1 quartil
primeiro_quartil = notas_vinhos.groupby(by='Tipo').quantile(0.25)
primeiro_quartil.loc['Total'] = notas_vinhos.quantile(0.25)
primeiro_quartil 
```

```python
#2 quartil
segundo_quartil = notas_vinhos.groupby(by='Tipo').quantile(0.5)
segundo_quartil.loc['Total'] = notas_vinhos.quantile(0.5)
segundo_quartil 
```

```python
#3 quartil
terceiro_quartil = notas_vinhos.groupby(by='Tipo').quantile(0.75)
terceiro_quartil.loc['Total'] = notas_vinhos.quantile(0.75)
terceiro_quartil 
```

## MEDIDAS DE DISPERSÃO

```python
#Mínimo
minimo_vinhos = notas_vinhos.groupby(by='Tipo').min()
minimo_vinhos.loc['Total'] = notas_vinhos.min()
minimo_vinhos
```

```python
#Máximo
maximo_vinhos = notas_vinhos.groupby(by='Tipo').max()
maximo_vinhos.loc['Total'] = notas_vinhos.max()
maximo_vinhos
```

```python
#Amplitude
amplitude_vinhos = notas_vinhos.groupby(by='Tipo').max() - notas_vinhos.groupby(by='Tipo').min() 
amplitude_vinhos.loc['Total'] = notas_vinhos.iloc[:,0:12].max() - notas_vinhos.iloc[:,0:12].min()
amplitude_vinhos
```

```python
#Intervalo Inter Quartil IQR
IQR = terceiro_quartil - primeiro_quartil
IQR
```

```python
#Adjacente inferior
Ai = primeiro_quartil - 1.5*IQR
Ai
```

```python
#Adjacente Superior
As = terceiro_quartil + 1.5*IQR
As
```

```python
#Desvio padrão
desvio_padrao_vinhos = notas_vinhos.groupby(by='Tipo').std() 
desvio_padrao_vinhos.loc['Total'] = notas_vinhos.std()
desvio_padrao_vinhos
```

```python
#Relação entre desvio padrão e média em porcentagem 
#(Para entender o tamanho do desvio padrão normalizado)
por_des_med = desvio_padrao_vinhos/media_vinhos*100
por_des_med
```

## COMPARAÇÃO DE MODELOS COM E SEM OUTLIERS

Para fazer o treinamento e teste dos modelos, separaremos os dados pela metade: 70% para treinar o modelo e 30% para testar

```python
#Regressão linear com outliers
treino = 0.7 #treinamento com 70% dos dados

quant_Verm = round(len(notas_vinhos[notas_vinhos['Tipo']=='Vermelho'])*treino)
quant_Branco = round(len(notas_vinhos[notas_vinhos['Tipo']=='Branco'])*treino)
```

**Separação de dados para treinamento e teste**

```python
#Vermelho
    #Dados para treinamento
X_com_Out_Verm_train = notas_vinhos[notas_vinhos['Tipo']=='Vermelho'].iloc[:quant_Verm,:-2]
y_com_Out_Verm_train = notas_vinhos[notas_vinhos['Tipo']=='Vermelho']['quality'].iloc[:quant_Verm]
    #Dados para teste
X_com_Out_Verm_test = notas_vinhos[notas_vinhos['Tipo']=='Vermelho'].iloc[quant_Verm:,:-2]
y_com_Out_Verm_test = notas_vinhos[notas_vinhos['Tipo']=='Vermelho']['quality'].iloc[quant_Verm:]

#Branco
    #Dados para treinamento
X_com_Out_Branco_train = notas_vinhos[notas_vinhos['Tipo']=='Branco'].iloc[:quant_Branco,:-2]
y_com_Out_Branco_train = notas_vinhos[notas_vinhos['Tipo']=='Branco']['quality'].iloc[:quant_Branco]
    #Dados para teste
X_com_Out_Branco_test = notas_vinhos[notas_vinhos['Tipo']=='Branco'].iloc[quant_Branco:,:-2]
y_com_Out_Branco_test = notas_vinhos[notas_vinhos['Tipo']=='Branco']['quality'].iloc[quant_Branco:]
```

**Treinamento e predição**

```python
LR_com_Out_Verm = LinearRegression()
LR_com_Out_Verm.fit(X_com_Out_Verm_train, y_com_Out_Verm_train) #ajuste dos parametros beta (vermelho)
y_pred_com_Out_Verm = LR_com_Out_Verm.predict(X_com_Out_Verm_test) #predicão de y com os 30% de dados restantes (vermelho)

LR_com_Out_Branco = LinearRegression()
LR_com_Out_Branco.fit(X_com_Out_Branco_train , y_com_Out_Branco_train) #ajuste dos parametros beta (branco)
y_pred_com_Out_Branco = LR_com_Out_Branco.predict(X_com_Out_Branco_test) #predicão de y com os 30% de dados restantes (branco)

#Exemplo gerado
y_pred_com_Out_Verm[:10]
```

Podemos notar que o modelo de regressão linear não prevê resultados por categoria. Nesse caso deveriam ser valores discretos de qualidade, porém o modelo previu valores continuos gerando as notas como 6,36

**Medida de explicação do modelo  $ R^{2}$**

```python
r2_score(y_com_Out_Verm_test , y_pred_com_Out_Verm) #R2 para o modelo de predição de vinho vermelho
```

```python
r2_score(y_com_Out_Branco_test , y_pred_com_Out_Branco) #R2 para o modelo de predição de vinho branco
```

Os $R^{2}$ tanto para o vinho vermelho quanto para o vinho branco são muito baixos. Então o próximo passo é remover os outliers esperando que esse indicador seja melhor.

#### REMOVENDO TODOS OS OUTLIERS DE TODAS AS VARIÁVEIS DE ENTRADA

A remoção dos outliers será feita apenas nas variáveis de entrada, pois remover outliers da saída nesse caso signifcaria excluir valores de qualidade dos vinhos, e queremos que o modelo seja capaz de prever dos piores aos melhores vinhos.

```python
vinho_vermelho_sem_out = vinho_vermelho
vinho_branco_sem_out = vinho_branco

#As é o Dataframe de adjacente superior e Ai é o Dataframe de Adjacente inferior criados no início do relatório

for coluna in caracteristicas_vinhos[:-1]:  #não considera a ultima variável, pois ela é a variável de saída
    vinho_vermelho_sem_out = vinho_vermelho_sem_out[vinho_vermelho_sem_out[coluna] < As[coluna].loc['Vermelho']]
    vinho_vermelho_sem_out = vinho_vermelho_sem_out[vinho_vermelho_sem_out[coluna] > Ai[coluna].loc['Vermelho']]
    vinho_branco_sem_out = vinho_branco_sem_out[vinho_branco_sem_out[coluna] < As[coluna].loc['Branco']]
    vinho_branco_sem_out = vinho_branco_sem_out[vinho_branco_sem_out[coluna] > Ai[coluna].loc['Branco']]
    
notas_vinhos_sem_out = pd.concat([vinho_vermelho_sem_out,vinho_branco_sem_out])
```

```python
plt.figure(figsize=[15.5,45])
n = 1

for caracteristica in caracteristicas_vinhos:
    plt.subplot(12,1,n) # 12 linhas, 1 coluna, 1o gráfico
    sns.boxplot(data=notas_vinhos_sem_out , x=caracteristica, y='Tipo', palette=['#b11226','#ffca70'])
    n += 1
```

Após a remoção dos outliers inicial, apareceram novos outliers devido à nova distribuição dos dados. Por serem poucos ouliers em relação ao caso passado, optamos por mantê los. 

#### CRIAÇÃO DO MODELO SEM OUTLIERS

```python
#Regressão linear sem outliers
quant_Verm_sem_out = round(len(notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Vermelho'])*treino)
quant_Branco_sem_out = round(len(notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Branco'])*treino)
```

**Separação de dados para treinamento e teste**

```python
#Vermelho
    #Dados para treinamento
X_sem_Out_Verm_train = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Vermelho'].iloc[:quant_Verm_sem_out,:-2]
y_sem_Out_Verm_train = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Vermelho']['quality'].iloc[:quant_Verm_sem_out]
    #Dados para teste
X_sem_Out_Verm_test = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Vermelho'].iloc[quant_Verm_sem_out:,:-2]
y_sem_Out_Verm_test = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Vermelho']['quality'].iloc[quant_Verm_sem_out:]

#Branco
    #Dados para treinamento
X_sem_Out_Branco_train = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Branco'].iloc[:quant_Branco_sem_out,:-2]
y_sem_Out_Branco_train = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Branco']['quality'].iloc[:quant_Branco_sem_out]
    #Dados para teste
X_sem_Out_Branco_test = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Branco'].iloc[quant_Branco_sem_out:,:-2]
y_sem_Out_Branco_test = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Branco']['quality'].iloc[quant_Branco_sem_out:]
```

**Treinamento e predição**

```python
LR_sem_Out_Verm = LinearRegression()
LR_sem_Out_Verm.fit(X_sem_Out_Verm_train, y_sem_Out_Verm_train)
y_pred_sem_Out_Verm = LR_sem_Out_Verm.predict(X_sem_Out_Verm_test)

LR_sem_Out_Branco = LinearRegression()
LR_sem_Out_Branco.fit(X_sem_Out_Branco_train , y_sem_Out_Branco_train)
y_pred_sem_Out_Branco = LR_sem_Out_Verm.predict(X_sem_Out_Branco_test)
```

**Medida de explicação do modelo  $ R^{2}$**

```python
r2_score(y_sem_Out_Verm_test , y_pred_sem_Out_Verm)
```

```python
r2_score(y_sem_Out_Branco_test , y_pred_sem_Out_Branco)
```

O $R^{2}$ para o modelo vermelho melhorou muito pouco, já para o vinho branco piorou muito o que já era ruim.
Então vamos fazer a análise do p-value para as variáveis de entrada (verificar se algum beta pode ser considerado zero, e se for remover a variável daquele beta)   
Serão removidas todas as variáveis com p-value maior que 5%

## ALGORITMO DE PREVISÃO POR REGRESSÃO LINEAR

Foi construido o estimador de regressão linear multipla e o preditor na mesma classe.

Como a conclusão da Etapa 1 é de que o modelo correto para estimar a qualidade dos vinhos provavelmente não é linear. Não faz sentido tentar utilizar apenas as variáveis com p-value menor que 5% para fazer uma boa previsão de qualidade. O R2 da análise após a remoção das variáveis com p-value alto mostrou que a predição continuará ruim.

Então o que fizemos foi comparar os betas gerados pelo algoritmo criado com os betas do pacote LinearRegression. Também comparamos o y predito com o pacote LinearRegression com o y predito do algoritmo criado.

```python
#classe da predição
class reg_lin_mult:
    
    def __init__(self):
        self.betas = []
        self.y_pred = []

    def fit(self, dados, saida):
        y = dados[saida]
        x = dados.drop(saida, axis=1)

        n = len(x)         #quantidade de pontos
        m = len(x.columns) #quantidade de variáveis explicativas

        X = np.zeros((m+1,m+1))
        Y = np.zeros(m+1)

        #para a primeira equação do sistema linear
        X[0,0] = n
        for i in range(n):
            Y[0] += y.iloc[i]
            for j in range(m):
                X[0,j+1] += x.iloc[i,j]    

        #para as equações restantes
        for k in range(m):
            for i in range(n):
                X[k+1,0] += x.iloc[i,k]
                Y[k+1] += x.iloc[i,k]*y.iloc[i]
                for j in range(m):
                    X[k+1,j+1] += x.iloc[i,k]*x.iloc[i,j]
                    
        self.betas = np.linalg.solve(X,Y)

    def pred(self,dados):
        if len(dados.columns) + 1 == len(self.betas):
            self.y_pred = []
            n = len(dados)
            m = len(dados.columns)
            
            for i in range(n):
                self.y_pred.append(self.betas[0])
                for j in range(m):
                    self.y_pred[i] += self.betas[j+1]*dados.iloc[i,j]
                self.y_pred[i] = self.y_pred[i].round(10)
                
        else:
            print('A quantidade de variáveis independentes deve ser igual à quantidade utilizada para treinamento do modelo')

            
```

**FIT**

A função fit() faz o somatorio das variaveis de entrada para construir o sistema de equações lineares gerados pela derivada = 0.  
Além disso, essa função organiza os valores em forma matricial X*b = Y
Onde X é uma matriz quadrada de tamanho igual à quantidade de betas, b é o vetor de betas a serem encontrados e Y é o vetor de saída da derivada = 0. Ao final usa-se linalg.solve(X,Y) para resolver o sistema linear e encontrar os betas

**PRED**

A função pred() recebe os dados de entrada, faz o somatório da multiplicação dos betas com as variáveis + B0 e cria um vetor com a saídas previstas

```python
#Exemplo de teste do algorimtmo:
# Os mesmos dados usados para o treinamento do modelo vermelho sem outliers usando o pacote LinearRegression()
# X_sem_Out_Verm_train = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Vermelho'].iloc[:quant_Verm_sem_out,:-2]
# y_sem_Out_Verm_train = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Vermelho']['quality'].iloc[:quant_Verm_sem_out]
#Os mesmos dados serão usados para ajustar o modelo e depois prever a saída

regressao1 = reg_lin_mult()

#apenas uma pequena mudança na parte iloc, pois temos que incluir a coluna 'quality' para funcionar o DataFrame no algoritmo
X_sem_Out_Verm_train2 = notas_vinhos_sem_out[notas_vinhos_sem_out['Tipo']=='Vermelho'].iloc[:quant_Verm_sem_out,:-1]

regressao1.fit(X_sem_Out_Verm_train2, 'quality')
```

## ALGORITMO DE PREVISÃO POR REGRESSÃO LOGISTÍCA

```python
from sklearn.linear_model import LogisticRegression
```

### Classificar os vinhos em bom ou ruim se é >= 6 ou < 6, respectivamente

```python
#cria a coluna Bom, coloca 0 e 1 de acordo com a qualidade
notas_vinhos_log = notas_vinhos.copy()
notas_vinhos_log = pd.get_dummies(notas_vinhos_log)
notas_vinhos_log['Bom'] = 0 #cria a coluna preenchendo tudo com zero
notas_vinhos_log['Bom'].loc[notas_vinhos_log['quality'] >= 6] = 1 #coloca 1 para valores de qualidade maiores ou iguais a 6

#Como os primeiros registros do dataframe são todos do tipo vermelho e os ultimos são todos do tipo branco, 
# precisa-se embraralhar os registros antes de separar em dados para treinamento e dados para teste do modelo
notas_vinhos_log = notas_vinhos_log.sample(frac=1, random_state=1)

#remove a coluna quality, pois ela não faz mais sentido para esse algortimo
qualidade = notas_vinhos_log['quality']
notas_vinhos_log.drop(['quality'], axis=1, inplace=True)
```

```python
#separa dados para treino e teste
treino = 0.7 # 70% de dados para treino
qtdade_treino = round(len(notas_vinhos_log)*treino)

x_treino = notas_vinhos_log.iloc[:qtdade_treino,:-1]
y_treino = notas_vinhos_log['Bom'].iloc[:qtdade_treino]

x_teste = notas_vinhos_log.iloc[qtdade_treino:,:-1]
y_teste = notas_vinhos_log['Bom'].iloc[qtdade_treino:]
```

```python
model = LogisticRegression() #instanciando a função
model.fit(x_treino, y_treino) #encontra os betas do modelo
print(model.intercept_, model.coef_)
```

```python
# funcão que define se a saída é 1 ou 0 baseado na probabilidade de corte

def define_cat(prob,corte):
    if prob >= 0 and prob <= 1 and corte >= 0 and corte <= 1: 
        if prob >= corte:
            return 1
        else:
            return 0
    else:
        return print('As probabilidades devem estar entre 0 e 1')
```

```python
#Aqui é feito a multiplicação de cada beta por sua respectiva variável
#Os valores são guardados no DataFrame betas_x
betas_x = pd.DataFrame([])

for n in range(len(x_teste.columns)-1):
    beta_x = pd.Series(x_teste.iloc[:,n].apply(lambda x: x*model.coef_[0,n]), name='b'+str(n+1)+'x'+str(n+1))
    betas_x = pd.concat([betas_x,beta_x], axis=1)

betas_x['b0'] = model.intercept_[0]

#z é o somatório dos betas multiplicados pelas variaveis para cada registro
z = betas_x.sum(axis=1)
z
```

```python
# Finalmente o Y_pred é feito através da função sigmoid, onde x recebe os valores z chegando a um probabilidade,
# a partir dessa probabilidade é definido se o vinho é bom ou ruim com a probabilidade de corte

Y_pred = z.apply(lambda x : define_cat(1/(1+np.exp(-x)),0.5))
Y_pred.name = 'y_pred'

# coloca-se o y_teste e o Y_pred no mesmo DataFrame
y_saida = pd.concat([y_teste,Y_pred],axis=1)
y_saida
```

### Comparação de resultados
Usa-se o y_teste separado anteriormente para comparar com a previsão do algoritmo

```python
# compara resultados entre o predito e a qualidade real do vinho e cria uma coluna com os valores correto ou errado
y_saida.loc[y_saida['Bom'] == y_saida['y_pred'], 'resultado'] = 'Correto' 
y_saida.loc[y_saida['Bom'] != y_saida['y_pred'], 'resultado'] = 'Errado'

#quantidade de valores corretos e errados
contagem_resultados = y_saida.groupby(by='resultado')['resultado'].count()
contagem_resultados
```
